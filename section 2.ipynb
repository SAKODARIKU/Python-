```python
def isiterable(obj):
    try:
        iter(obj) #イテラブルとは要素を1つずつ順番に取り出せる性質を持っているという意味
        return True
    except TypeError:
        return False
```

```python
def add_and_maybe_multiple(a,b,c=None):#cは何も入れなければ初期値Noneという意味
    result=a+b
    if c is not None:
        result=result*c
    return result
```

```python
add_and_maybe_multiple(2,4,"f")
```

```python
add_and_maybe_multiple(2,4)
```

```python
sequence=[1,2,None,4,None,5]
total=0
for value in sequence:
    if value is None:
        continue    #要素にNoneがあったら次に進むという意味
    total+=value
print(total)
```

```python
sequence=[1,2,0,4,6,5,2,1]
total_until_5=0
for value in sequence:
    if value == 5:
        break
    total_until_5+=value
print(total_until_5)
```
