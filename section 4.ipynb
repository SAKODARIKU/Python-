```python
np.diag(X) #対角要素または非対角要素　一次元の場合はそれが対角要素の対角行列を返す
```

```python
np.trace(X) #diagの要素の和
```

```python
det(X)
```

```python
eig(X) #固有値及び固有ベクトルを返す
```

```python
qr(X) #QR分解　直行行列Qと上三角行列Rに分解　逆行列をdot(inv(R),Q.T)で求められる
```

```python
pinv(X) #ムーア-ベンローズの疑似逆行列を返す Aを入力した行列とすると、ABA=Bとなる行列Bを返す
```

```python
svd(X) #特異値分解(USVhの積に)した結果を返す　U,Vhは直行行列、Sは対角行列でSの対角成分を特異値と呼ぶ
```

```python
nwalks=5000
```

```python
nsteps=1000
```

```python
draws=rng.integers(0,2,size=(nwalks,nsteps))
```

```python
steps=np.where(draws>0,1,-1)
```

```python
walks=steps.cumsum(axis=1)
```

```python
walks
```

```python
walks.max()
```

```python
walks.min()
```

```python
hits30=(np.abs(walks)>=30).any(axis=1)
```

```python
hits30
```

```python
crossing_times=(np.abs(walks[hits30])>=30).argmax(axis=1)
```

```python
crossing_times
```
