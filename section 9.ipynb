```python
from datetime import datetime
import matplotlib.dates as dates
import matplotlib.font_manager as fm
fm.get_font_names()
```

```python
font_options={"family":"MS Gothic"}
plt.rc("font",**font_options)

fig,ax=plt.subplots()

data=pd.read_csv("spx.csv",index_col=0,parse_dates=True)
spx=data["SPX"]

spx.plot(ax=ax,color="black")

crisis_data=[
    (datetime(2007,10,11),"上昇相場のピーク"),
    (datetime(2008,3,12),"ベア・スターンズ危機"),
    (datetime(2008,9,15),"リーマン破綻")
]

for date,label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
               xytext=(date,spx.asof(date)+225),
               arrowprops=dict(facecolor="black",headwidth=4,width=2,
                              headlength=4),
               horizontalalignment="left",verticalalignment="top")
    
datefmt=dates.DateFormatter("%Y年%m月")
ax.xaxis.set_major_formatter(datefmt)

ax.set_xlabel(None)

ax.set_xlim(["1/1/2007","1/1/2011"])
ax.set_ylim([600,1800])

ax.set_title("2008～2009年の金融危機の重要な日付")
```

```python
tips=pd.read_csv("tips.csv")
```

```python
tips.head()
```

```python
party_counts=pd.crosstab(tips["day"],tips["size"])
```

```python
party_counts=party_counts.reindex(index=["Thur","Fri","Sat","Sun"])
```

```python
party_counts
```

```python
party_counts=party_counts.loc[:,2:5]
```

```python
party_pcts=party_counts.div(party_counts.sum(axis="columns"),axis="index")
```

```python
party_pcts
```

```python
party_pcts.plot.bar(stacked=True)
```

```python
macro=pd.read_csv("macrodata.csv")
```

```python
data=macro[["cpi","m1","tbilrate","unemp"]]
```

```python
trans_data=np.log(data).diff().dropna()
#diff()は前後のデータの差
```

```python
trans_data.tail()
```

```python
ax=sns.regplot(x="m1",y="unemp",data=trans_data)
ax.title("Change in log(m1) versus log(unemp)")
```

```python
sns.pairplot(trans_data,diag_kind="kde",plot_kws={"alpha":0.2})
```

```python
sns.catplot(x="day",y="tip_pct",hue="time",col="smoker",kind="bar",data=tips[tips.tip_pct<1])
```

```python
sns.catplot(x="day",y="tip_pct",row="time",col="smoker",kind="bar",
           data=tips[tips.tip_pct<1])
```
